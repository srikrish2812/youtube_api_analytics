
> 2.b  Provide the following statistics for query string ‚Äúavatar movie‚Äù of top 50 videos sorted by relevance in the US region ( 7 marks )

> Output expected: video ID, title, no of views, no of likes,no of comments exported to CSV file






Reference link: https://developers.google.com/youtube/v3/docs/videos/list

```python
# < code block > 
#2.b


import csv
data_row_list = []

heading_fields = ['Video ID','Title','No of Views','No of Likes', 'No of Comments']

def get_statistics(video_id):
    video_stats_request = youtube.videos().list(
        part='snippet,statistics',
        id = video_id,
    )
    video_stats_response = video_stats_request.execute()
    row_as_dict = {}
    item = video_stats_response['items'][0]
    video_id = item['id']
    title = item['snippet']['title']
    total_views = int(item['statistics']['viewCount'])
    total_likes = int(item['statistics']['likeCount'])
    total_comments = int(item['statistics']['commentCount'])
    row_as_dict['Video ID'] = video_id
    row_as_dict['Title'] = title
    row_as_dict['No of Views'] = total_views
    row_as_dict['No of Likes'] = total_likes
    row_as_dict['No of Comments'] = total_comments
    data_row_list.append(row_as_dict)
    print(row_as_dict)

#making request
search_request = youtube.search().list(
    q='avatar movie',
    part='id,snippet',
    regionCode = 'US',
    order='relevance',
    type='video',
    maxResults=50
)
search_response = search_request.execute()
for item in search_response['items']:
    video_id = item['id']['videoId']
    get_statistics(video_id)

#print(data_row_list)


with open('stats_top_50.csv','w') as csv_file:
    writer_instance = csv.DictWriter(csv_file,fieldnames=heading_fields)
    writer_instance.writeheader()
    writer_instance.writerows(data_row_list)
    print("EXPORTED TO CSV FILE stats_top_50.csv")
```

Execute this request and print the response using the pprint() method.
 - Note: pprint() will print out the response from the API in a more human-readable format.
- Reference link:  [link](https://developers.google.com/youtube/v3/docs )


> **Output expected** : Use the python library ‚Äú pprint ‚Äú to print the output of the program with the following properties  etag, items, id , kind, snippet and snippet to have the text display field which represents the comment of videos.






```python
# < code block >
#3.b Print response using pprint :

import pprint
formatted_responses_list = []
for video in top_10_comments_data:

    video_id = video['Video ID']
    print("\n\n")
    comments_request = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
        )
    comments_response = comments_request.execute()
    formatted_response = pprint.pformat(comments_response)
    formatted_responses_list.append(formatted_response)
    pprint.pprint(comments_response)
```
